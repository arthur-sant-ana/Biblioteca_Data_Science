### Converting .tif images to .png (fastai does not accept .tif) (Done in setup)

### Setting up libraries to construct segmentation model

```python
from fastai.vision.all import *
```

```python
path = Path('AWS_files')
path_imgs = path/'images_png'
path_masks = path/'masks'
```

### Creating DataBlock with png images as features and masks as labels

```python
def label_func(fn):
    return path_masks/f"{fn.stem}_mask.tif"
```

```python
codes = [0, 1]  # Background = 0, Building = 1

dblock = DataBlock(
    blocks=(ImageBlock, MaskBlock(codes=codes)),
    get_items=get_image_files,
    get_y=label_func,
    splitter=RandomSplitter(),
    item_tfms=Resize(256),  # or 512 depending on GPU
    batch_tfms=aug_transforms()
)
```

```python
dls = dblock.dataloaders(path_imgs, bs=16)
```

```python
learn = unet_learner(dls, resnet34, metrics=Dice())
learn.fine_tune(7)
```

### Visualizing results

```python
learn.show_results()
plt.show()
```

### Converting test .tif files into .png files (done in setup)

### Using model to segment pictures from testing bach

```python
img = PILImage.create("AWS_files/test_images_png/RGB-PanSharpen_AOI_3_Paris_img177.png")
pred_mask, _, probs = learn.predict(img)
```

```python
print(pred_mask.shape)
print(np.unique(np.array(pred_mask)))
```

```python
resized_img = img.resize(pred_mask.size())

plt.figure(figsize=(6,6))
plt.imshow(resized_img)
plt.imshow(np.array(pred_mask)*256, alpha=0.3, cmap='Reds')
plt.title("Overlay: Prediction on Image")
plt.axis('off')
plt.show()
```
