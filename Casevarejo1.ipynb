```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_excel('varejo.xlsx')
df.head()
```

```python
df2 = pd.read_excel('cliente_varejo.xlsx')
df2.head()
```

```python
df.info()
```

```python
df['estado'].fillna('MS', inplace = True)
```

```python
df.info()
```

```python
df['Preço'].fillna(df['Preço_com_frete'], inplace = True)
```

```python
df[df['estado'] == 'MS']
```

```python
df_filtrado = df[df['Preço'] <= df['Preço_com_frete']]
df_filtrado.info()
```

Pergunta #1; Departamentos mais vendidos

```python
df_1 = df.groupby(['Nome_Departamento'])['idcompra'].count().sort_values(ascending = False).reset_index()
df_1.head()
```

```python
sns.barplot(data = df_1, x = 'Nome_Departamento', y = 'idcompra')
```

Pergunta #2; Média de preço com frete por departamento

```python
df_2 = df.groupby(['Nome_Departamento'])['Preço_com_frete'].agg('mean').sort_values(ascending = False).reset_index()
df_2.head()
```

```python
sns.barplot(data = df_2, x = 'Nome_Departamento', y = 'Preço_com_frete')
```

Pergunta #3; Quatidade de vendas por mês

```python
df['Mes'] = df['Data'].dt.month
df.sort_values(by = 'Data')
```

```python
df_3 = df.groupby(['Mes'])['idcompra'].count().sort_values(ascending = False).reset_index()
df_3.head(12)
```

```python
sns.barplot(data = df_3, x = 'Mes', y = 'idcompra')
```

Pergunta #4; Media de renda para cada tipo de canal

Merging Duas bases de dados

```python
df2.groupby(['cliente_Log']).count()
```

```python
df3 = df.merge(df2, how = 'left', on = 'cliente_Log')
df3.head()
```

```python
df3 = df3.replace('APP', 'Aplicativo')
```

```python
df_4 = df3.groupby(['idcanalvenda'])['renda'].agg('mean').reset_index()
df_4.head()
```

```python
sns.barplot(data = df_4, x = 'idcanalvenda', y = 'renda')
```

Pergunta #5; Media de idade de clientes por bandeira

```python
df_5 = df3.groupby(['bandeira'])['idade'].agg('mean').reset_index()
df_5.head()
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize = (7, 4))

plt.bar(df_5['bandeira'], df_5['idade'], color = 'blue')
plt.title('Média de idade por bandeira')
plt.xlabel('Bandeira')
plt.ylabel('Idade')
```

```python
df_6 = df3.groupby(['Data'])['idcompra'].count().reset_index()
df_6.head()
```

```python
fig = plt.figure(figsize = (10, 8))

plt.plot(df_6['Data'], df_6['idcompra'], color = 'red')
plt.title('Vendas por dia')
plt.xlabel('Data')
plt.ylabel('Vendas')
```

```python
fig = plt.figure(figsize = (10, 8))\

plt.bar(df_3['Mes'], df_3['idcompra'], color = 'green')
plt.title('Vendas por mês')
plt.xlabel('Mês')
plt.ylabel('Vendas')
```
